```python
path = 'D:\\iDX\\dataset\\voda.pdf'
```

```python
import fitz
from operator import itemgetter
import pandas as pd
import json
```

```python
doc=fitz.open(path)
```

```python
blocks = doc.loadPage(8).getText('blocks')
```

```python
blocks = sorted(blocks , key= itemgetter(1,0))
```

```python
df = pd.DataFrame(blocks)
```

```python
for i,j in df.iterrows():
    print(i,'==',j[4])
```

```python
df = df.loc[df.index[8:17]]
```

```python
json = []
for i,j in df.iterrows():
    b_type = input(f'{j[4]}')
    json.append({'BoundingBox':[{'Left':j[0],
                         'Top':j[1],
                         'Width':j[2]-j[0],
                         'Height':j[3]-j[1]}],
         'Type':'Block',
         'Block_No':j[5],
         'Text':j[4],
         'BlockType':b_type,
         'Child':[]})
```

```python
json_ = json
```

```python
with open('voda.json',mode='w') as file_:
    json.dump(json_,file_)
```

```python
words = doc.loadPage(8).getText('words')
```

```python
blocks_list = [24,33,34,35,36,37,39,38,23]
```

```python
words_main = []
for i in blocks_list:
    words_ = list(filter(lambda x:x[5]==i,words))
    for j in words_:
        words_main.append({
        'BoundingBox':[{'Left':j[0],
                         'Top':j[1],
                         'Width':j[2]-j[0],
                         'Height':j[3]-j[1]}],
         'Type':'Word',
         'Text':j[4],
        'Id': f'block{j[5]}_line{j[6]}_word{j[7]}'
        })
```
