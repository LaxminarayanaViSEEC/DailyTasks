```python
import pandas as pd
import numpy as np
```

```python
def block_merge_determiner(previous_block, current_block):
    prev_y2 = previous_block[3]
    curr_y1 = current_block[1]
    difference = curr_y1 - prev_y2
    if difference <= 2.5:
        return True
    else:
        return False
```

```python
block = pd.read_csv('pdfExtraction/block.csv')
line =  pd.read_csv('pdfExtraction/line.csv')
word =  pd.read_csv('pdfExtraction/word.csv')
```

```python
block = block.sort_values(by=['bbox_top', 'bbox_left'])
block.reset_index(inplace=True,drop=True)
```

```python
for index,row in block.iterrows():
    if index==0:
        prev_index = index
        prev_block = (row['bbox_left'],
                     row['bbox_top'],
                     row['bbox_right'],
                     row['bbox_bottom'])
    else:
        curr_index = index
        curr_block = (row['bbox_left'],
                     row['bbox_top'],
                     row['bbox_right'],
                     row['bbox_bottom'])
        if block_merge_determiner(prev_block,curr_block):
            
            #lets update coordinates
            block.loc[prev_index,'bbox_left'] = min(block.loc[prev_index,'bbox_left'],row['bbox_left'])
            block.loc[prev_index,'bbox_top'] = min(block.loc[prev_index,'bbox_top'],row['bbox_top'])
            block.loc[prev_index,'bbox_right'] = max(block.loc[prev_index,'bbox_right'],row['bbox_right'])
            block.loc[prev_index,'bbox_bottom'] = max(block.loc[prev_index,'bbox_bottom'],row['bbox_bottom'])
            
            #lets update the size
            block.loc[prev_index,'size'] = list(block.loc[prev_index,'size']).extend(list(row['size']))
            
            #lets update the text
            block.loc[prev_index,'text'] = f"{block.loc[prev_index,'text']} {row['text']}"
            
            #lets update the font_name
            block.loc[prev_index,'font_name'] = list(block.loc[prev_index,'font_name']).extend(list(row['font_name']))
            
            #finally drop the row from block
            block.drop(index=curr_index, inplace = True)
            
        else:
            prev_index = curr_index
            prev_block = curr_block
```
